```python
from moexalgo import Market, Ticker
import pandas as pd
import time
```

```python
# Акции SBER
sber = Ticker('SBER')

# Все акции
stocks = Market('stocks')
```

```python
start_date = '2020-01-01'
end_date = '2023-12-20'

date_array = pd.date_range(start=start_date, end=end_date, freq='D')
date_array
```

```python
tradestats = pd.DataFrame()
for d in date_array:
    print(d.strftime('%Y-%m-%d'))
    tradestats = pd.concat([tradestats, sber.candles(date=d.strftime('%Y-%m-%d'), till_date=d.strftime('%Y-%m-%d'), period='1h')])
    # tradestats = pd.concat([tradestats, sber.candles(date=d, till_date=d, period='1m')])
    time.sleep(0.5)
```

```python
tradestats.to_csv('tradestats_hour.csv', index=None)
```
